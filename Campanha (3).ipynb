# 1 Análise da MYK Marketing

## 1.1 Cenário: Avaliando o Engajamento das Redes Sociais

Nos últimos meses, a equipe da MYK Marketing, especializada em campanhas digitais para grandes marcas de moda, beleza e lifestyle, percebeu um comportamento instável no engajamento das campanhas nas redes sociais. Enquanto alguns posts viralizavam organicamente, outros com investimentos similares tinham desempenho fraco.

A diretoria comercial, pressionada por metas de ROI, lançou uma missão clara para o time de dados:

**“Precisamos entender, com dados, o que realmente impulsiona o engajamento das nossas campanhas e como podemos prever o sucesso de novos conteúdos.”**

A partir disso, o time de analytics iniciou um projeto de investigação estratégica, conectadas por um único objetivo: transformar dados de engajamento em decisões mais inteligentes e lucrativas.

## 1.2 Objeto do Projeto

Maximizar o retorno sobre investimento (ROI) nas campanhas digitais da MYK Marketing por meio da análise de dados de engajamento em diferentes plataformas, considerando conteúdo, região e uso de hashtags e para isso devem responder as perguntas abaixo:

**1)Quais formatos de postagens geram o maior volume de engajamento absoluto?**


**2)Quais combinações de conteúdo, região e hashtag maximizam o engajamento?**


**3)Quais sinergias entre tipo de conteúdo e região devem ser exploradas?**


**4)Qual o valor médio de engajamento por hashtag, controlando por plataforma e região?**


**5)Quais características se repetem entre os posts de alto engajamento?**



**Premissa do Projeto**

Este estudo tem como foco principal a análise de engajamento nas campanhas digitais, entendendo que o engajamento é um indicador antecipado de performance e influência direta no ROI.

Embora o retorno financeiro (ROI) não tenha sido mensurado diretamente neste projeto, ele será abordado em uma próxima etapa.

A proposta atual visa identificar padrões de alto engajamento, os quais serão base para estratégias que maximizem o ROI em futuras campanhas.

A literatura de marketing digital e casos anteriores demonstram que campanhas com maior engajamento têm mais chance de conversão em receita.

Assim, os padrões identificados aqui como o uso de vídeos, hashtags específicas e foco regional têm potencial de elevar o ROI, sendo recomendados para testes em campanhas futuras com acompanhamento de KPIs financeiros.



## 1.3 Análise SWOT

**Forças (Strengths)**





Especialização da equipe: A MYK Marketing possui um time de analytics dedicado, com experiência em campanhas digitais para marcas de moda, beleza e lifestyle, o que garante conhecimento profundo do mercado.



Foco em dados: O projeto é orientado por uma abordagem data-driven, utilizando métricas de engajamento para embasar decisões, o que aumenta a precisão das estratégias.



Objetivos claros: As cinco perguntas do projeto são bem definidas, cobrindo desde a identificação de padrões até a previsão de resultados, alinhando-se diretamente com a meta de maximizar o ROI.



Diversidade de plataformas analisadas: A análise abrange diferentes redes sociais, permitindo uma visão ampla e comparativa do desempenho de conteúdos.



Capacidade de adaptação: A possibilidade de identificar combinações específicas de conteúdo, região e hashtags permite personalizar campanhas para maximizar o engajamento.

**Fraquezas (Weaknesses)**





Instabilidade no desempenho: A inconsistência no engajamento (alguns posts viralizam, outros têm desempenho fraco) indica lacunas no entendimento dos fatores que impulsionam o sucesso.



Dependência de dados históricos: A análise depende da qualidade e da quantidade de dados disponíveis, o que pode limitar a identificação de padrões em plataformas ou regiões menos exploradas.



Complexidade operacional: Responder às cinco perguntas exige integrar dados de múltiplas variáveis (plataforma, região, hashtag, tipo de conteúdo), o que pode demandar tempo e recursos significativos.



Falta de clareza sobre custos: O projeto foca no engajamento ajustado por views, mas não menciona explicitamente como os custos de produção de conteúdo serão incorporados nas análises de ROI.



Risco de generalização: Há o perigo de assumir que padrões identificados em uma plataforma ou região se apliquem universalmente, o que pode levar a decisões menos eficazes.

**Oportunidades (Opportunities)**





Personalização de campanhas: A identificação de combinações eficazes de conteúdo, região e hashtags pode permitir campanhas altamente segmentadas, aumentando o engajamento e o ROI.



Previsão de resultados: Desenvolver modelos preditivos para antecipar o engajamento de novos posts pode otimizar a alocação de recursos e priorizar conteúdos com maior potencial.



Expansão geográfica: Explorar sinergias entre tipo de conteúdo e região pode revelar mercados regionais subaproveitados, permitindo escalar campanhas bem-sucedidas.



Otimização de hashtags: Identificar hashtags de alto desempenho ajustadas por plataforma e região pode melhorar a eficiência das campanhas sem aumento significativo de custos.



Tendências de mercado: O foco em moda, beleza e lifestyle alinha-se com setores em crescimento, oferecendo oportunidades para capitalizar em tendências emergentes nas redes sociais.

**Ameaças (Threats)**





Mudanças nas plataformas: Alterações em algoritmos de redes sociais podem impactar o engajamento, dificultando a aplicação de padrões identificados no projeto.



Concorrência acirrada: O mercado de marketing digital é altamente competitivo, e outras agências podem adotar estratégias semelhantes, reduzindo a vantagem competitiva da MYK.



Saturação de conteúdo: A grande quantidade de conteúdo nas redes sociais pode dificultar a viralização orgânica, mesmo com combinações otimizadas de hashtags e formatos.



Custos crescentes: Investimentos em campanhas de alto engajamento (como vídeos ou parcerias com influenciadores) podem aumentar, pressionando o ROI se os resultados não forem proporcionais.



Riscos de privacidade e regulamentação: Novas regras de privacidade ou restrições em redes sociais podem limitar o acesso a dados de engajamento, impactando a capacidade de análise.

```python
import matplotlib.pyplot as plt

# Dados
plataformas = ['YouTube', 'TikTok', 'Instagram', 'Twitter']
valores = [1324, 1260, 1212, 1204]

# Calculando porcentagens
total = sum(valores)
porcentagens = [round((v / total) * 100, 1) for v in valores]  # Arredondado com 1 casa decimal

# Cores oficiais das plataformas
cores = ['#FF0000', '#000000', '#E4405F', '#1DA1F2']  # YouTube, TikTok, Instagram, Twitter

# Criando o gráfico de barras horizontal
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
barras = plt.barh(plataformas, porcentagens, color=cores)  # Cria as barras horizontais

# Adicionando título e rótulos dos eixos
plt.title('Distribuição Percentual das Plataformas de Mídia Social', fontsize=16)
plt.xlabel('Porcentagem (%)', fontsize=14)
plt.ylabel('Plataforma', fontsize=14)

# Ajustando o limite do eixo X para criar mais espaço à direita
plt.xlim(0, max(porcentagens) + 5)  # Adiciona 5% extra de margem à direita

# Adicionando porcentagens ao lado de cada barra
for i, v in enumerate(porcentagens):
    plt.text(v + 0.3, i, f'{v}%', va='center', fontsize=12, ha='left')

# Ajustando layout
plt.tight_layout()  # Ajusta o layout para evitar cortes

# Exibindo o gráfico
plt.show()
```

**Interpretação**

*YouTube é a plataforma mais popular*: Com 26,5%, o YouTube lidera como a plataforma de mídia social mais utilizada entre as opções apresentadas. Isso pode refletir a preferência por conteúdo em vídeo mais longo e diversificado.

*Instagram e Twitter estão praticamente empatados*: Instagram (24,2%) e Twitter (24,1%) têm percentuais muito próximos, indicando que ambas as plataformas são igualmente populares, mas com propósitos distintos (Instagram focado em imagens e vídeos, enquanto o Twitter é mais textual).

*Diferença mínima entre as plataformas*: A diferença entre a plataforma mais popular (YouTube) e a menos popular (Twitter) é de apenas 2,4%, sugerindo que todas as plataformas têm um nível de adesão semelhante, com pequenas variações.

*Diversidade de preferências*: O gráfico mostra que não há uma plataforma dominante, o que indica que o público está distribuído de forma relativamente equilibrada entre as opções, refletindo a diversidade de interesses e comportamentos nas redes sociais.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Carregar o conjunto de dados
df = pd.read_csv("Viral_Social_Media_Trends.csv")

# ============== 1. COMPARAÇÃO DE MÉDIAS POR PLATAFORMA E TIPO DE CONTEÚDO ==============
# Calcular métricas de engajamento por 1000 visualizações
df['Curtidas_por_1000_visualizacoes'] = (df['Likes'] / df['Views']) * 1000
df['Comentarios_por_1000_visualizacoes'] = (df['Comments'] / df['Views']) * 1000
df['Compartilhamentos_por_1000_visualizacoes'] = (df['Shares'] / df['Views']) * 1000

# Criar uma pontuação de engajamento composta (média ponderada)
df['Pontuacao_Engajamento'] = (df['Curtidas_por_1000_visualizacoes'] * 0.4 + 
                               df['Comentarios_por_1000_visualizacoes'] * 0.4 + 
                               df['Compartilhamentos_por_1000_visualizacoes'] * 0.2)

# Médias por plataforma e tipo de conteúdo
numeric_features = ['Curtidas_por_1000_visualizacoes', 'Comentarios_por_1000_visualizacoes', 
                    'Compartilhamentos_por_1000_visualizacoes', 'Pontuacao_Engajamento']

# Visualização das médias
plt.figure(figsize=(12, 6))
engagement_means = df.groupby(['Platform', 'Content_Type'])[numeric_features].mean().round(2)
engagement_means['Pontuacao_Engajamento'].unstack().plot(kind='bar', rot=0, 
                                                         title='Médias da Pontuação de Engajamento por Plataforma e Tipo de Conteúdo')
plt.ylabel('Pontuação de Engajamento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('medias_engajamento.png')
plt.show()

# Boxplot
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
sns.boxplot(x='Content_Type', y='Pontuacao_Engajamento', hue='Platform', data=df)
plt.title('Distribuição da Pontuação de Engajamento por Tipo de Conteúdo e Plataforma')
plt.xticks(rotation=45)

# Heatmap
plt.subplot(2, 1, 2)
pivot_table = df.groupby(['Platform', 'Content_Type'])['Pontuacao_Engajamento'].mean().unstack()
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Heatmap da Pontuação de Engajamento por Plataforma e Tipo de Conteúdo')

plt.tight_layout()
plt.savefig('visualizacoes_engajamento.png')
plt.show()

# ============== 2. EXIBIR RESULTADOS NO CONSOLE ==============
# Exibir informações iniciais do dataset

print("\n🔹 Médias das variáveis de engajamento por plataforma e tipo de conteúdo:")
print(engagement_means)

# ANOVA para verificar diferenças significativas
anova_results = []
for feature in numeric_features:
    groups = [group[feature].values for name, group in df.groupby('Content_Type')]
    f_stat, p_val = f_oneway(*groups)
    anova_results.append({
        'Feature': feature,
        'F_Statistic': f_stat,
        'P_Value': p_val,
        'Significant': p_val < 0.05
    })

anova_df = pd.DataFrame(anova_results)
print("\n✅ Resultados do teste ANOVA (diferenças entre tipos de conteúdo):")
print(anova_df)

# ============== 3. RELATÓRIO FINAL ==============
print("\n📌 CONCLUSÕES SOBRE O ENGAJAMENTO EM MÍDIAS SOCIAIS")
print("=====================================================================================")
print("A análise revelou os seguintes padrões:\n")

# Variáveis com diferenças significativas
significant_anova = anova_df[anova_df['Significant']]['Feature'].tolist()
print(f"🔹 Variáveis com diferenças estatísticas significativas entre tipos de conteúdo: {', '.join(significant_anova)}")

# Identificar tipos de conteúdo com maior variação
high_diff_vars = engagement_means.std().sort_values(ascending=False).head(3).index.tolist()
print(f"\n🔹 Variáveis com maior variação entre plataformas e conteúdos: {', '.join(high_diff_vars)}")

# Encontrar o melhor tipo de conteúdo por plataforma
top_content_by_platform = engagement_means['Pontuacao_Engajamento'].unstack().idxmax(axis=1).to_dict()
print("\n🔹 Melhores tipos de conteúdo por plataforma:")
for platform, content_type in top_content_by_platform.items():
    score = engagement_means.loc[(platform, content_type), 'Pontuacao_Engajamento']
    print(f"  - {platform}: {content_type} (Pontuação: {score:.2f})")

# Recomendações finais
print("\n💡 Recomendações para aumentar o ROI das campanhas:")
print("- Priorize a criação de conteúdos com maior pontuação de engajamento, como vídeos no Instagram ou imagens no Twitter.")
print("- Aloque maior orçamento para plataformas com altos índices de engajamento para maximizar o retorno.")
print("- Implemente testes A/B com os tipos de conteúdo identificados como mais eficazes para otimizar o desempenho.")
print("- Monitore continuamente as métricas de engajamento para ajustar estratégias em tempo real.")
```

**B)Quais combinações de conteúdo, região e hashtag maximizam o engajamento?**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from scipy.stats import chi2_contingency
import warnings
warnings.filterwarnings('ignore')

# Carregar os dados
df = pd.read_csv("Viral_Social_Media_Trends.csv")

# Definir engajamento alto (top 25% de pontuação)
df['Engagement_Score'] = (df['Likes'] + df['Comments'] + df['Shares']) / df['Views']
high_engagement_threshold = df['Engagement_Score'].quantile(0.75)
df['High_Engagement'] = (df['Engagement_Score'] >= high_engagement_threshold).astype(int)


print(df[['Views', 'Likes', 'Shares', 'Comments', 'Engagement_Score']].describe())

# ============== 1. ANÁLISE EXPLORATÓRIA ==============
print("\n🔹 Distribuição de engajamento alto por categoria:")
engagement_dist = df.groupby(['Content_Type', 'Region', 'Platform'])['High_Engagement'].mean().unstack().fillna(0)
print(engagement_dist)

# Visualização
plt.figure(figsize=(12, 8))
sns.heatmap(engagement_dist, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Taxa de Engajamento Alto por Combinação de Fatores')
plt.tight_layout()
plt.show()

# ============== 2. MODELAGEM PREDITIVA ==============
# Pré-processamento
le = LabelEncoder()
df['Content_Type_Encoded'] = le.fit_transform(df['Content_Type'])
df['Region_Encoded'] = le.fit_transform(df['Region'])
df['Platform_Encoded'] = le.fit_transform(df['Platform'])
df['Hashtag_Encoded'] = le.fit_transform(df['Hashtag'])

features = ['Content_Type_Encoded', 'Region_Encoded', 'Platform_Encoded', 'Hashtag_Encoded']
X = df[features]
y = df['High_Engagement']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Decision Tree
dt = DecisionTreeClassifier(max_depth=4, random_state=42)
dt.fit(X_train, y_train)

# Feature Importance
importance_dt = pd.DataFrame({
    'Feature': features,
    'Importance_DT': dt.feature_importances_
}).sort_values('Importance_DT', ascending=False)

print("\n✅ Importância das variáveis (Decision Tree):")
print(importance_dt)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

importance_rf = pd.DataFrame({
    'Feature': features,
    'Importance_RF': rf.feature_importances_
}).sort_values('Importance_RF', ascending=False)

print("\n✅ Importância das variáveis (Random Forest):")
print(importance_rf)

# ============== 3. ANÁLISE DAS COMBINAÇÕES IDEIAS ==============
# Obter combinações com maior probabilidade de engajamento alto
def get_top_combinations(df, n=5):
    comb_df = df.groupby(['Content_Type', 'Region', 'Hashtag', 'Platform'])['High_Engagement'] \
               .agg(['mean', 'count']).reset_index()
    comb_df = comb_df[comb_df['count'] >= 10]  # Filtro por mínimo de ocorrências
    return comb_df.sort_values('mean', ascending=False).head(n)

top_combinations = get_top_combinations(df)
print("\n🏆 Top combinações para engajamento alto:")
print(top_combinations)

# ============== 4. TESTES ESTATÍSTICOS ==============
# Teste qui-quadrado para associação entre variáveis categóricas e engajamento
chi2_results = []
for feature in ['Content_Type', 'Region', 'Hashtag']:
    contingency_table = pd.crosstab(df[feature], df['High_Engagement'])
    chi2, p, dof, _ = chi2_contingency(contingency_table)
    n = contingency_table.values.sum()
    v_cramer = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
    chi2_results.append({
        'Feature': feature,
        'Chi2': chi2,
        'P_Value': p,
        'Cramer_V': v_cramer,
        'Significant': p < 0.05
    })

chi2_df = pd.DataFrame(chi2_results)
print("\n✅ Resultados do teste Qui-Quadrado (associação com engajamento alto):")
print(chi2_df[['Feature', 'Chi2', 'P_Value', 'Cramer_V', 'Significant']])

# ============== 5. RELATÓRIO FINAL ==============
print("\n📌 CONCLUSÕES SOBRE COMBINAÇÕES DE ALTO ENGAJAMENTO")
print("=====================================================================================")
print("A análise revelou os seguintes padrões:\n")

# Variáveis mais importantes
top_features = importance_rf.sort_values('Importance_RF', ascending=False).head(3)['Feature'].tolist()
print(f"🔹 Variáveis mais importantes para prever engajamento alto: {', '.join(top_features)}")

# Combinações de sucesso
print("\n🔹 Combinações com maior taxa de engajamento alto:")
for i, row in top_combinations.iterrows():
    print(f"  - {row['Content_Type']} + {row['Region']} + {row['Hashtag']} + {row['Platform']} (Taxa: {row['mean']:.2%})")

# Associações estatísticas
significant_chi2 = chi2_df[chi2_df['Significant']]['Feature'].tolist()
if significant_chi2:
    print(f"\n🔹 Variáveis com associação estatística significativa com engajamento alto: {', '.join(significant_chi2)}")
else:
    print("\n🔹 Nenhuma variável categórica mostrou associação estatística forte com engajamento alto.")

# Recomendações finais
print("\n💡 Recomendações para campanhas futuras:")
print("- Priorize as combinações de conteúdo-região-hashtag identificadas como mais eficazes")
print("- Otimize o tipo de conteúdo de acordo com a plataforma e região de destino")
print("- Utilize as hashtags que demonstraram maior correlação com engajamento alto")
print("- Monitore continuamente o desempenho para identificar novas tendências")
print("- Realize testes A/B com as combinações recomendadas para validação")
```

```python
import geopandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from statsmodels.formula.api import ols
import warnings
warnings.filterwarnings('ignore')

# Carregar os dados
df = pd.read_csv("Viral_Social_Media_Trends.csv")

# Calcular métricas de desempenho
df['ROI'] = (df['Likes'] + df['Shares']*2 + df['Comments']*3) / df['Views']  # Métrica customizada de retorno
df['Engagement_Rate'] = (df['Likes'] + df['Comments'] + df['Shares']) / df['Views']

print("📊 Informações iniciais do dataset:")
print(df[['Views', 'Likes', 'Shares', 'Comments', 'ROI']].describe())

# ============== 1. ANÁLISE DE INTERAÇÃO CONTEÚDO-REGIÃO ==============
# Regressão com termo de interação
print("\n🔹 Análise de regressão para sinergias conteúdo-região:")
model = ols('ROI ~ Content_Type * Region', data=df).fit()
interaction_effects = pd.DataFrame(model.params[model.params.index.str.contains(':')])
interaction_effects.columns = ['Coeficiente']
print(interaction_effects.sort_values('Coeficiente', ascending=False))

# Visualização das interações
plt.figure(figsize=(14, 7))
sns.pointplot(data=df, x='Region', y='ROI', hue='Content_Type', 
              ci=95, dodge=True, palette='viridis')
plt.title('Interação entre Tipo de Conteúdo e Região (ROI Médio)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ============== 2. CLUSTERIZAÇÃO DE MERCADOS ==============
# Preparar dados para clustering
cluster_data = df.pivot_table(index='Region', 
                             columns='Content_Type', 
                             values='ROI', 
                             aggfunc='mean').fillna(0)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)
cluster_data['Cluster'] = cluster_labels

print("\n✅ Grupos de regiões com padrões similares de desempenho:")
print(cluster_data['Cluster'].value_counts())

# Visualização dos clusters
plt.figure(figsize=(10, 6))
sns.heatmap(cluster_data.drop('Cluster', axis=1), 
            cmap='YlGnBu', 
            annot=True, 
            fmt=".2f")
plt.title('Padrões de ROI por Região e Tipo de Conteúdo')
plt.tight_layout()
plt.show()

# ============== 3. MAPAS GEOGRÁFICOS (OPCIONAL) ==============
try:
    import geopandas as gpd
    # Simulação de dados geográficos (substitua por seus dados reais)
    geo_data = df.groupby('Region')['ROI'].mean().reset_index()
    geo_data['Latitude'] = np.random.uniform(-30, 30, len(geo_data))  # Dummy
    geo_data['Longitude'] = np.random.uniform(-60, 60, len(geo_data)) # Dummy
    
    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=geo_data, x='Longitude', y='Latitude', 
                   size='ROI', hue='ROI', sizes=(50, 200), palette='viridis')
    plt.title('Desempenho por Região (Tamanho = ROI)')
    for line in range(0, geo_data.shape[0]):
        plt.text(geo_data.Longitude[line]+1, 
                 geo_data.Latitude[line], 
                 geo_data.Region[line], 
                 horizontalalignment='left')
    plt.show()
    
except ImportError:
    print("\n⚠️ Geopandas não instalado. Visualização de mapas simplificada.")

# ============== 4. RELATÓRIO FINAL (CÓDIGO CORRIGIDO) ==============
print("\n📌 CONCLUSÕES SOBRE SINERGIAS CONTEÚDO-REGIÃO")
print("=====================================================================================")
print("A análise revelou os seguintes padrões:\n")

# Melhores interações (extração robusta)
top_interactions = interaction_effects.sort_values('Coeficiente', ascending=False).head(3)
print(f"🔹 Melhores sinergias conteúdo-região:")
for idx, row in top_interactions.iterrows():
    try:
        # Extrai conteúdo e região de formatos como "Content_Type[T.X]:Region[T.Y]"
        parts = idx.split(':')
        content = parts[0].split('[')[-1].replace('T.', '').strip(']')
        region = parts[1].split('[')[-1].replace('T.', '').strip(']')
        print(f"  - {content} + {region} (Efeito: {row['Coeficiente']:.4f})")
    except:
        continue

# Características dos clusters
print("\n🔹 Perfil dos grupos de regiões:")
for cluster in sorted(cluster_data['Cluster'].unique()):
    cluster_regions = cluster_data[cluster_data['Cluster'] == cluster].index.tolist()
    print(f"  Grupo {cluster}: {', '.join(cluster_regions)}")
    
    top_content = cluster_data.drop('Cluster', axis=1).loc[cluster_data['Cluster'] == cluster].mean().idxmax()
    print(f"    → Melhor tipo de conteúdo: {top_content}")

# Recomendações finais
print("\n💡 Recomendações estratégicas:")
print("- Priorizar as combinações conteúdo-região com maiores coeficientes positivos")
print("- Desenvolver estratégias específicas para cada cluster de regiões")
print("- Testar as sinergias identificadas em campanhas piloto antes de escalar")
print("- Monitorar continuamente o ROI por tipo de conteúdo em cada região")
print("- Adaptar o mix de conteúdo conforme a resposta regional")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Carregar o dataset
df = pd.read_csv("Viral_Social_Media_Trends.csv")

# ============== 1. ANÁLISE INICIAL DO DATASET ==============


# Verificar distribuição do Engagement_Level
print("\n🔹 Distribuição do Nível de Engajamento:")
print(df['Engagement_Level'].value_counts())

# Visualização da distribuição de Engagement_Level
plt.figure(figsize=(10, 6))
sns.countplot(x='Engagement_Level', data=df, order=['Low', 'Medium', 'High'])
plt.title('Distribuição do Nível de Engajamento', fontsize=14)
plt.ylabel('Contagem', fontsize=12)
plt.xlabel('Nível de Engajamento', fontsize=12)
plt.tight_layout()
plt.savefig('engagement_distribution.png', dpi=300)
plt.close()

# ============== 2. PRÉ-PROCESSAMENTO E ENGENHARIA DE FEATURES ==============
# Criar novas features relacionadas a engajamento
df['Likes_per_View'] = df['Likes'] / df['Views'].replace(0, 1)
df['Shares_per_View'] = df['Shares'] / df['Views'].replace(0, 1)
df['Comments_per_View'] = df['Comments'] / df['Views'].replace(0, 1)
df['Engagement_Rate'] = (df['Likes'] + df['Shares'] + df['Comments']) / df['Views'].replace(0, 1)

# Adicionar features de proporção
df['Engagement_per_View'] = df['Engagement_Rate'] / df['Views'].replace(0, 1)
df['Likes_Share_Ratio'] = df['Likes'] / df['Shares'].replace(0, 1)
df['Comments_Share_Ratio'] = df['Comments'] / df['Shares'].replace(0, 1)

# Codificar variáveis categóricas
categorical_features = ['Platform', 'Content_Type', 'Region', 'Hashtag']
df_encoded = pd.get_dummies(df, columns=categorical_features, drop_first=False)

# Selecionar features (excluir colunas não relevantes para modelagem)
features = [col for col in df_encoded.columns if col not in ['Post_ID', 'Engagement_Level', 'Engagement_Rate', 'Total_Interactions']]
X = df_encoded[features]
y = df_encoded['Engagement_Rate']

# ============== 3. VALIDAÇÃO CRUZADA ESTRATIFICADA ==============
# Criar folds estratificados baseados na distribuição do engajamento
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Lista para armazenar métricas de cada fold
mse_scores = []
r2_scores = []

# Realizar validação cruzada estratificada
for train_idx, val_idx in skf.split(X, pd.qcut(df['Engagement_Rate'], 5, labels=False, duplicates='drop')):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Escalar as features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_val_scaled = scaler.transform(X_val)
    
    # Treinar modelo para este fold
    model = RandomForestRegressor(
        n_estimators=200,
        max_depth=15,
        random_state=42,
        n_jobs=-1,
        min_samples_split=5,
        min_samples_leaf=2
    )
    
    model.fit(X_train_scaled, y_train)
    
    # Prever e calcular métricas
    y_pred = model.predict(X_val_scaled)
    mse_scores.append(mean_squared_error(y_val, y_pred))
    r2_scores.append(r2_score(y_val, y_pred))

# Calcular métricas médias
print("\n✅ Resultados do Modelo Random Forest Regressor (Validação Estratificada):")
print(f"Erro Quadrático Médio (MSE): {np.mean(mse_scores):.4f} (+/- {np.std(mse_scores):.4f})")
print(f"Coeficiente de Determinação (R²): {np.mean(r2_scores):.4f} (+/- {np.std(r2_scores):.4f})")

# ============== 4. TREINAMENTO DO MODELO FINAL ==============
# Dividir em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=pd.qcut(y, 5, labels=False, duplicates='drop'))

# Escalar features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Treinar o modelo final
model = RandomForestRegressor(
    n_estimators=200,
    max_depth=15,
    random_state=42,
    n_jobs=-1,
    min_samples_split=5,
    min_samples_leaf=2
)
model.fit(X_train_scaled, y_train)

# Prever no conjunto de teste
y_pred = model.predict(X_test_scaled)

# ============== 5. AVALIAÇÃO DO MODELO COMO CLASSIFICADOR ==============
# Discretizar previsões em Low, Medium, High
thresholds = np.percentile(y_train, [33.3, 66.6])
y_pred_classes = np.digitize(y_pred, thresholds)
y_test_classes = np.digitize(y_test, thresholds)

# Mapear valores numéricos para categorias
class_mapping = {0: 'Low', 1: 'Medium', 2: 'High'}
y_pred_labels = [class_mapping[val] for val in y_pred_classes]
y_test_labels = [class_mapping[val] for val in y_test_classes]

# Avaliar como classificação
print("\n✅ Resultados da Classificação (após discretização):")
print(classification_report(y_test_classes, y_pred_classes, target_names=['Low', 'Medium', 'High']))

# Matriz de confusão
cm = confusion_matrix(y_test_classes, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Low', 'Medium', 'High'],
            yticklabels=['Low', 'Medium', 'High'])
plt.title('Matriz de Confusão - Random Forest Regressor (Discretizado)', fontsize=14)
plt.xlabel('Previsto', fontsize=12)
plt.ylabel('Real', fontsize=12)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300)
plt.close()

# ============== 6. ANÁLISE DE IMPORTÂNCIA DAS FEATURES ==============
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': model.feature_importances_
}).sort_values('Importance', ascending=False)

# Filtrar features com importância maior que 0 e formatar para 2 casas decimais
feature_importance = feature_importance[feature_importance['Importance'] > 0]
feature_importance['Importance'] = feature_importance['Importance'].round(5)

print("\n🔹 Importância das Features:")
print(feature_importance)

# Visualizar importância das features (top 20)
plt.figure(figsize=(12, 10))
sns.barplot(x='Importance', y='Feature', data=feature_importance.head(20))
plt.title('Top 20 Features mais Importantes (Random Forest Regressor)', fontsize=14)
plt.xlabel('Importância', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300)
plt.close()

# ============== 7. VISUALIZAÇÃO DA IMPORTÂNCIA POR CATEGORIA ==============
# Separar features por categorias
engagement_metrics = ['Likes_per_View', 'Shares_per_View', 'Comments_per_View', 'Engagement_per_View', 
                       'Likes_Share_Ratio', 'Comments_Share_Ratio', 'Views', 'Likes', 'Shares', 'Comments']
platform_cols = [col for col in features if 'Platform_' in col]
content_cols = [col for col in features if 'Content_Type_' in col]
region_cols = [col for col in features if 'Region_' in col]
hashtag_cols = [col for col in features if 'Hashtag_' in col]

categories = {
    'Métricas de Engajamento': engagement_metrics,
    'Plataformas': platform_cols,
    'Tipos de Conteúdo': content_cols,
    'Regiões': region_cols,
    'Hashtags': hashtag_cols
}

# Criar visualização por categoria
fig, axes = plt.subplots(len(categories), 1, figsize=(14, 24))
fig.suptitle('Importância das Features por Categoria', fontsize=16)

for i, (category, cols) in enumerate(categories.items()):
    category_data = feature_importance[feature_importance['Feature'].isin(cols)].sort_values('Importance', ascending=False)
    
    if not category_data.empty:
        sns.barplot(x='Importance', y='Feature', data=category_data, ax=axes[i])
        axes[i].set_title(f'{category}', fontsize=14)
        axes[i].set_xlabel('Importância', fontsize=12)
        axes[i].set_ylabel('Feature', fontsize=12)
    else:
        axes[i].text(0.5, 0.5, f'Nenhuma feature de {category} encontrada', 
                    horizontalalignment='center', verticalalignment='center', fontsize=12)
        axes[i].set_title(f'{category}', fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig('importance_by_category.png', dpi=300)
plt.close()

print("=====================================================================================")

# ============== 8. ANÁLISE DE COMBINAÇÕES DE ALTO ENGAJAMENTO ==============
# Identificar combinações de alto engajamento
high_engagement_mask = df['Engagement_Level'] == 'High'
high_engagement_df = df[high_engagement_mask]

# Combinações de Platform, Content_Type, Region e Hashtag
combinations = high_engagement_df.groupby(['Platform', 'Content_Type', 'Region', 'Hashtag']).size().reset_index(name='count')
combinations = combinations.sort_values('count', ascending=False).head(30)

print("\n🔹 Top 30 Combinações de Alto Engajamento (Platform, Content_Type, Region, Hashtag):")
print(combinations)

# Visualizar top combinações
plt.figure(figsize=(14, 16))
sns.barplot(x='count', y=combinations.apply(lambda x: f"{x['Platform']} + {x['Content_Type']} + {x['Region']} + {x['Hashtag']}", axis=1), data=combinations)
plt.title('Top 30 Combinações para Alto Engajamento', fontsize=14)
plt.xlabel('Número de Posts com Alto Engajamento', fontsize=12)
plt.ylabel('Combinação (Plataforma + Tipo + Região + Hashtag)', fontsize=12)
plt.tight_layout()
plt.savefig('high_engagement_combinations.png', dpi=300)
plt.close()

print("=====================================================================================")

# ============== 9. RELATÓRIO FINAL ==============

print("=====================================================================================")

print("\n📌 CONCLUSÕES SOBRE A PREVISÃO DE ENGAJAMENTO")

print("A análise revelou os seguintes padrões:\n")

# Destacar features mais importantes
top_features = feature_importance['Feature'].head(3).tolist()
print(f"🔹 Features com maior impacto no engajamento: {', '.join(top_features)}")

# Explicação das principais variáveis de importância
print("\n🔹 Explicação das principais variáveis de importância:")

print("- Likes_per_View (0.40): Representa a proporção de curtidas em relação às visualizações de um post. Sua alta importância (0.40) indica que posts com maior proporção de curtidas por visualização são fortemente associados a maior engajamento.")

print("- Engagement_per_View (0.38): Calculada como a taxa de engajamento (soma de curtidas, compartilhamentos e comentários dividida por visualizações) dividida pelas visualizações. Com importância de 0.38, reflete a intensidade do engajamento relativo ao alcance do post, sendo uma métrica crítica.")

print("- Views (0.17): O número total de visualizações de um post. Sua importância de 0.17 sugere que o volume bruto de visualizações tem um impacto moderado na taxa de engajamento, mas é menos influente que métricas normalizadas.")

print("- Shares_per_View (0.02): Proporção de compartilhamentos em relação às visualizações. Sua baixa importância (0.02) indica que os compartilhamentos têm um impacto menor na previsão do engajamento comparado a curtidas ou engajamento geral.")

print("- Comments_per_View (0.02): Proporção de comentários em relação às visualizações. Com importância de 0.02, sugere que os comentários têm uma influência limitada na taxa de engajamento, similar aos compartilhamentos.")

print("- Content_Type_Live Stream: Indica se o tipo de conteúdo é uma transmissão ao vivo. Sua presença entre as features importantes sugere que transmissões ao vivo têm um impacto específico no engajamento, possivelmente devido à interação em tempo real.")


print("=====================================================================================")
# Desempenho do modelo
print("\n🔹 Desempenho do modelo:")
print(f"- O modelo Random Forest Regressor previu a taxa de engajamento com MSE: {np.mean(mse_scores):.4f} e R²: {np.mean(r2_scores):.4f}.")
print(f"- Acurácia média na validação estratificada (R²): {np.mean(r2_scores):.2f}.")


print("=====================================================================================")
# Insights por plataforma
platform_importance = feature_importance[feature_importance['Feature'].isin(platform_cols)]
if not platform_importance.empty:
    top_platform = platform_importance.iloc[0]['Feature'].replace('Platform_', '')
    print(f"\n🔺 Plataforma com maior influência no engajamento: {top_platform}")
else:
    print("\n🔺 Nenhuma plataforma específica destacou-se na análise.")



# Insights por tipo de conteúdo
content_importance = feature_importance[feature_importance['Feature'].isin(content_cols)]
if not content_importance.empty:
    top_content = content_importance.iloc[0]['Feature'].replace('Content_Type_', '')
    print(f"🔺 Tipo de conteúdo com maior influência no engajamento: {top_content}")
else:
    print(f"🔺 Nenhum tipo de conteúdo específico destacou-se na análise.")

# Insights por hashtag
hashtag_importance = feature_importance[feature_importance['Feature'].isin(hashtag_cols)]
if not hashtag_importance.empty:
    top_hashtag = hashtag_importance.iloc[0]['Feature'].replace('Hashtag_', '')
    print(f"🔺 Hashtag com maior influência no engajamento: {top_hashtag}")
else:
    print(f"🔺 Nenhuma hashtag específica destacou-se na análise.")

# Top combinações
if not combinations.empty:
    top_combo = combinations.iloc[0]
    print(f"🔺 Combinação com maior ocorrência em alto engajamento: {top_combo['Platform']} + {top_combo['Content_Type']} + {top_combo['Region']} + {top_combo['Hashtag']} ({top_combo['count']} posts)")


print("=====================================================================================")

# Recomendações finais
print("\n💡 Recomendações:")
print("- Priorizar posts com altas taxas de engajamento normalizadas (Likes/Views, Engagement/Views).")
print("- Investir nas combinações de plataforma, tipo de conteúdo, região e hashtag que demonstraram alto engajamento.")
print("- Usar o modelo treinado para prever a taxa de engajamento de novos posts e alocar recursos em campanhas com maior ROI.")
print("- Monitorar métricas em tempo real e ajustar estratégias com base nos resultados do modelo.")
print("- Explorar otimizações específicas para as combinações de maior performance identificadas na análise.")
print("- Considerar testar as combinações de alto engajamento em plataformas onde ainda não foram exploradas.")
```
